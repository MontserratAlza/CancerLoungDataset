## Estimados de locación y variabilidad

```python
#Estimados de locación: Se refieren a medidas de tendencia central, como la media, mediana, y moda, que describen el "centro" de los datos.
#Estimados de variabilidad: Son medidas que indican la dispersión de los datos, como la desviación estándar, varianza, y rango intercuartílico.
df_general = df_current.copy()

df_general.head()
```

```python
# Calcular la media y mediana (estimados de locación)
mean_age = df_general['age'].mean()
median_age = df_general['age'].median()

# Calcular la desviación estándar y rango intercuartílico (estimados de variabilidad)
std_age = df_general['age'].std()
iqr_age = df_general['age'].quantile(0.75) - df_general['age'].quantile(0.25)

print(f"Media de edad: {mean_age}, Mediana de edad: {median_age}")
print(f"Desviación estándar de edad: {std_age}, Rango intercuartílico de edad: {iqr_age}")
#Estos estimados son útiles para describir la edad promedio de los pacientes y cómo varía dentro del grupo.
#Valores típicos: Son aquellos que están cerca de la media o mediana y dentro del rango normal.
#Valores atípicos (outliers): Son datos que se desvían significativamente de los valores típicos, lo que puede influir en el análisis y los modelos de machine learning.
```

```python
# Calcular los límites para los valores atípicos basados en el IQR
Q1 = df_general['age'].quantile(0.25)
Q3 = df_general['age'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filtrar los valores atípicos
outliers = df_general[(df_general['age'] < lower_bound) | (df_general['age'] > upper_bound)]

print(f"Límite inferior: {lower_bound}, Límite superior: {upper_bound}")
print(df_general['age'].describe())

# Validar si hay valores atípicos
if outliers.empty:
    print("No se encontraron valores atípicos en la edad.")
else:
    print(f"Valores atípicos en la edad:\n{outliers}")

```

# Tema 5: Distribuciones Muestrales y Técnicas de Evaluación de Modelos

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import mean_squared_error
import numpy as np
import matplotlib.pyplot as plt

# Descargar el dataset
url = 'https://drive.google.com/uc?id=19dBuqEFjc0zx1quYLLnlJLRGyeWP8GC7'
output = 'dataset_cancer.csv'
gdown.download(url, output, quiet=False)

# Cargar el dataset
dataset_cancer = pd.read_csv(output)

# Convertir columnas categóricas a valores booleanos (1 y 0)
dataset_cancer['GENDER'] = dataset_cancer['GENDER'].map({'M': 1, 'F': 0})

# Lista de columnas a convertir a valores booleanos
columns_to_convert = ['SMOKING', 'YELLOW_FINGERS', 'ANXIETY', 'PEER_PRESSURE',
                      'CHRONIC_DISEASE', 'FATIGUE', 'ALLERGY', 'WHEEZING',
                      'ALCOHOL_CONSUMING', 'COUGHING', 'SHORTNESS_OF_BREATH',
                      'SWALLOWING_DIFFICULTY', 'CHEST_PAIN']

# Convertir columna objetivo 'LUNG_CANCER' a booleano
dataset_cancer['LUNG_CANCER'] = dataset_cancer['LUNG_CANCER'].map({'YES': 1, 'NO': 0})

# Convertir las columnas especificadas a valores booleanos
for column in columns_to_convert:
    dataset_cancer[column] = dataset_cancer[column].map({'Yes': 1, 'No': 0})

# 1. Función de Bootstrap para calcular intervalos de confianza de una estadística
def bootstrap_statistic_confidence_interval(data, stat_function, n_bootstrap=1000, confidence_level=0.95):
    bootstrap_statistics = []
    for _ in range(n_bootstrap):
        # Crear una muestra con reemplazo
        sample = data.sample(frac=1, replace=True)
        # Calcular la estadística de interés
        bootstrap_statistics.append(stat_function(sample))

    # Calcular el promedio estimado y el intervalo de confianza
    stat_estimate = np.mean(bootstrap_statistics)
    ci_lower = np.percentile(bootstrap_statistics, (1 - confidence_level) / 2 * 100)
    ci_upper = np.percentile(bootstrap_statistics, (1 + confidence_level) / 2 * 100)
    return stat_estimate, ci_lower, ci_upper

# Ejemplo 1: Bootstrap para la media de 'AGE'
age_data = dataset_cancer['AGE']
age_mean, age_ci_lower, age_ci_upper = bootstrap_statistic_confidence_interval(age_data, np.mean)

# Ejemplo 2: Bootstrap para la proporción de fumadores
smoking_data = dataset_cancer['SMOKING']
smoking_proportion, smoking_ci_lower, smoking_ci_upper = bootstrap_statistic_confidence_interval(smoking_data, np.mean)

# Visualizar distribuciones muestrales de la edad y proporción de fumadores usando bootstrap
plt.hist([age_data.sample(frac=1, replace=True).mean() for _ in range(1000)], bins=30, alpha=0.5, label='Distribución Bootstrap - Edad (AGE)')
plt.axvline(age_mean, color='r', linestyle='--', label=f'Media de edad: {age_mean:.2f}')
plt.axvline(age_ci_lower, color='g', linestyle=':', label=f'IC 95% inferior: {age_ci_lower:.2f}')
plt.axvline(age_ci_upper, color='b', linestyle=':', label=f'IC 95% superior: {age_ci_upper:.2f}')
plt.legend()
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.title('Distribución muestral de la media de la Edad usando Bootstrap')
plt.show()

# 2. Preparación de datos para la Regresión Lineal Múltiple
# Separar las características y la variable objetivo
X = dataset_cancer.drop(columns=['LUNG_CANCER'])
y = dataset_cancer['LUNG_CANCER']

# Ejemplo 3: Dividir el dataset en entrenamiento y prueba (30% y 50%)
for test_size in [0.3, 0.5]:
    print(f"\nDivisión de datos con {int(test_size*100)}% como prueba:")
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=42)

    # Entrenar el modelo
    model = LinearRegression()
    model.fit(X_train, y_train)

    # Evaluación en el conjunto de prueba
    y_pred = model.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    print(f"Error cuadrático medio (MSE) en el conjunto de prueba: {mse:.4f}")

# Ejemplo 4: Validación cruzada con diferentes valores de k (k=5 y k=10)
for k in [5, 10]:
    print(f"\nValidación cruzada con k={k}:")
    cv_scores = cross_val_score(model, X, y, cv=k, scoring='neg_mean_squared_error')
    cv_mean = -cv_scores.mean()
    cv_std = cv_scores.std()
    print(f"MSE promedio: {cv_mean:.4f}")
    print(f"Desviación estándar del MSE: {cv_std:.4f}")

# Mostrar resultados del bootstrap y la proporción de fumadores
print("\nResultados de Bootstrap:")
print("Media estimada de AGE:", age_mean)
print("Intervalo de confianza 95% de AGE:", (age_ci_lower, age_ci_upper))
print("Proporción estimada de fumadores:", smoking_proportion)
print("Intervalo de confianza 95% de la proporción de fumadores:", (smoking_ci_lower, smoking_ci_upper))
```

# Tema 8 Rubrica: Introducción a Machine Learning: Clasificación No Supervisada y Supervisada


```python
import pandas as pd
import gdown
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans

# Descargar el dataset
url = 'https://drive.google.com/uc?id=19dBuqEFjc0zx1quYLLnlJLRGyeWP8GC7'
output = 'dataset_cancer.csv'
gdown.download(url, output, quiet=False)

# Cargar el dataset
df = pd.read_csv(output)

# Convertir "Yes" y "No" a 1 y 0 en las columnas especificadas y asegurar que el tipo sea entero
yes_no_columns = [
    'SMOKING', 'YELLOW_FINGERS', 'ANXIETY', 'PEER_PRESSURE',
    'CHRONIC_DISEASE', 'FATIGUE', 'ALLERGY', 'WHEEZING',
    'ALCOHOL_CONSUMING', 'COUGHING', 'SHORTNESS_OF_BREATH',
    'SWALLOWING_DIFFICULTY', 'CHEST_PAIN', 'LUNG_CANCER'
]

# Convertir todos los valores a minúsculas y quitar espacios extra
df[yes_no_columns] = df[yes_no_columns].apply(lambda col: col.str.strip().str.lower())

# Convertir "yes" a 1 y "no" a 0
df[yes_no_columns] = df[yes_no_columns].replace({'yes': 1, 'no': 0}).astype(int)

# Seleccionar características para clustering
features = df[['AGE', 'SMOKING', 'FATIGUE']].copy()

# Aplicar K-Means con el número de clusters deseado
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(features)

# Agregar las etiquetas de cluster al DataFrame
df['Cluster'] = kmeans.labels_

# Visualización de clusters
plt.figure(figsize=(10, 6))
sns.scatterplot(x='AGE', y='FATIGUE', hue='Cluster', data=df, palette='Set1', s=100)
plt.title('Clusters de Fumadores por Edad y Fatiga')
plt.xlabel('Edad')
plt.ylabel('Fatiga (0: No, 1: Sí)')
plt.legend(title='Cluster')
plt.grid(True)
plt.show()
```

```python
# Seleccionar características para clustering: SMOKING y YELLOW_FINGERS
features = df[['SMOKING', 'YELLOW_FINGERS']]

# Aplicar K-Means
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(features)

# Agregar las etiquetas de cluster al DataFrame
df['Cluster'] = kmeans.labels_

# Visualizar los clusters
plt.figure(figsize=(10, 6))
sns.scatterplot(x='SMOKING', y='YELLOW_FINGERS', hue='Cluster', data=df, palette='Set1', s=100)
plt.title('Clusters por Fumar y Dedos Amarillos')
plt.xlabel('Fumar (0: No, 1: Sí)')
plt.ylabel('Dedos Amarillos (0: No, 1: Sí)')
plt.legend(title='Cluster')
plt.grid(True)
plt.show()
```

```python
# Seleccionar características para clustering: ALCOHOL_CONSUMING y CHRONIC_DISEASE
features = df[['ALCOHOL_CONSUMING', 'CHRONIC_DISEASE']]

# Aplicar K-Means
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(features)

# Agregar las etiquetas de cluster al DataFrame
df['Cluster'] = kmeans.labels_

# Visualizar los clusters
plt.figure(figsize=(10, 6))
sns.scatterplot(x='ALCOHOL_CONSUMING', y='CHRONIC_DISEASE', hue='Cluster', data=df, palette='Set1', s=100)
plt.title('Clusters de Consumo de Alcohol y Enfermedades Crónicas')
plt.xlabel('Consumo de Alcohol (0: No, 1: Sí)')
plt.ylabel('Enfermedad Crónica (0: No, 1: Sí)')
plt.legend(title='Cluster')
plt.grid(True)
plt.show()
```
